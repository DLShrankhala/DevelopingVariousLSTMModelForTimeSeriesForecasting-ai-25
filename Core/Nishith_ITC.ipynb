```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

data = pd.read_csv('ITC.csv')
data
```

```python
# Plot the data
plt.figure(figsize=(10, 5))
plt.plot(data['Date'], data['Open'], marker='o')

# Add titles and labels
plt.title('Open Value Over Time')
plt.xlabel('Date')
plt.ylabel('Open Value')

# Improve the date formatting on the x-axis
plt.gcf().autofmt_xdate()

# Show the plot
plt.show()
```
